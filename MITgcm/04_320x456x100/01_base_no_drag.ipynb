### Load bathymetry variables

```python
with Dataset(bathy_file1, 'r') as nbl:
    lat = nbl.variables['Y'][:]
    lon = nbl.variables['X'][:]
    Z = nbl.variables['Z'][:]
    Zl = nbl.variables['Zl'][:]
    depth = nbl.variables['Depth'][:]
    # centers mask
    hFacC = nbl.variables['HFacC'][:]
    hfac = np.ma.masked_values(hFacC, 0)
    mask = np.ma.getmask(hfac)
    # U mask
    hFacW = nbl.variables['HFacW'][:]
    hfacW = np.ma.masked_values(hFacW, 0)
    maskU = np.ma.getmask(hfacW)
    # V mask
    hFacS = nbl.variables['HFacS'][:]
    hfacS = np.ma.masked_values(hFacS, 0)
    maskV = np.ma.getmask(hfacS)
    # masked depth
    dep_masked = np.ma.masked_values(depth,0)
    mask_dep = np.ma.getmask(dep_masked)
    
with Dataset(state_file1, 'r') as nbl:
    Zind = (nbl.variables['diag_levels'][:]).astype(int)
```

```python
fig, ax = plt.subplots(1,1,figsize=(9,7))
ax.set_facecolor('tan')
pc = ax.pcolormesh(lon,lat,np.ma.masked_array(depth, mask=mask_dep), cmap=cmo.cm.deep_r)
cb = plt.colorbar(pc)
cn2 = ax.contour(lon,lat,depth, colors='0.5', levels=[250])

lon_ind = 255
lat_ind1 = 70
lat_ind2 = 120

# ssh transects
ax.plot(lon[:],lat[lat_ind1]*np.ones_like(lon),'r-')
ax.plot(lon[:],lat[lat_ind2]*np.ones_like(lon),'r-')
ax.plot(lon[:],lat[lat_ind2+50]*np.ones_like(lon),'r-')
ax.plot(lon[:],lat[lat_ind2+100]*np.ones_like(lon),'r-')
ax.plot(lon[:],lat[lat_ind2+150]*np.ones_like(lon),'r-')
ax.plot(lon[:],lat[lat_ind2+200]*np.ones_like(lon),'r-')
ax.plot(lon[:],lat[lat_ind2+250]*np.ones_like(lon),'r-')
ax.plot(lon[:],lat[lat_ind2+300]*np.ones_like(lon),'r-')

# sponge widths
ax.plot(lon[12]*np.ones_like(lat),lat[:],'-', color='orange')
ax.plot(lon[:],lat[12]*np.ones_like(lon),'-', color='orange')
ax.plot(lon[:],lat[-12]*np.ones_like(lon),'-', color='orange')

cb.set_label('depth / m')
ax.set_xlabel('lon')
ax.set_ylabel('lat')

ax.set_aspect(1)
```

### Difference among slip, no slip and slip w/ quad drag on surface oscillations

```python
eta_file2 = '/Volumes/KYUUBI/SVB/04_320x456x100/02_w_drag/etaGlob.nc'
eta_file3 = '/Volumes/KYUUBI/SVB/04_320x456x100/03_no_slip/etaGlob.nc'

with Dataset(eta_file2, 'r') as nbl:
    etab1 = nbl.variables['ETAN'][:,:,lat_ind1,-12]
    etab2 = nbl.variables['ETAN'][:,:,lat_ind2,-30]
    etab3 = nbl.variables['ETAN'][:,:,lat_ind2+50,-104]
    etab4 = nbl.variables['ETAN'][:,:,lat_ind2+100,-139]
    etab5 = nbl.variables['ETAN'][:,:,lat_ind2+150,-154]
    etab6 = nbl.variables['ETAN'][:,:,lat_ind2+200,-190]
    etab7 = nbl.variables['ETAN'][:,:,lat_ind2+250,-211]
    etab8 = nbl.variables['ETAN'][:,:,lat_ind2+300,-251]

with Dataset(eta_file3, 'r') as nbl:
    etac1 = nbl.variables['ETAN'][:,:,lat_ind1,-12]
    etac2 = nbl.variables['ETAN'][:,:,lat_ind2,-30]
    etac3 = nbl.variables['ETAN'][:,:,lat_ind2+50,-104]
    etac4 = nbl.variables['ETAN'][:,:,lat_ind2+100,-139]
    etac5 = nbl.variables['ETAN'][:,:,lat_ind2+150,-154]
    etac6 = nbl.variables['ETAN'][:,:,lat_ind2+200,-190]
    etac7 = nbl.variables['ETAN'][:,:,lat_ind2+250,-211]
    etac8 = nbl.variables['ETAN'][:,:,lat_ind2+300,-251]
    
etas = [eta1,eta2,eta3,eta4,eta5,eta6,eta7,eta8]
etasb = [etab1,etab2,etab3,etab4,etab5,etab6,etab7,etab8]
etasc = [etac1,etac2,etac3,etac4,etac5,etac6,etac7,etac8]
```

```python
## Run 01 - Run 02
fig, (ax1) = plt.subplots(1,1, figsize=(7,4), sharey=True,sharex=True)

for eta, etab, ind, off in zip(etas, etasb, inds, range(len(etas))):
    ax1.plot(time/3600,(0.5*off)+(eta-etab)*100, label=r'%1.1f$^{\circ}$ N' %lat[ind])

for ii in range(5):
    ax1.axvline((5.*ii), color='0.5', alpha=0.4)
ax1.set_ylabel('$\eta$ / cm')
ax1.set_xlabel('time')
ax1.set_xlim(0,24)
ax1.legend(bbox_to_anchor=(1,1))
ax1.set_title('slip run - no slip run')
```
